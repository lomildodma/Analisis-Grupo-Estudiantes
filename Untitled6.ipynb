# Análisis de Datos del Grupo de IA

```python
# Importa la librería pandas para manejo de datos
import pandas as pd

# Carga el archivo CSV 'Estudiantes.csv' en un DataFrame llamado 'estudiantes'
estudiantes = pd.read_csv("datasets/Estudiantes.csv")

# Muestra las primeras filas del DataFrame para revisar su contenido
estudiantes
```

```python
# Toma la variable 'Edad' y 'Promedio' del DataFrame 'estudiantes' y aplica el método .describe()
estudiantes[['Edad', 'Promedio']].describe()
```

```python
# Toma la variable 'Genero' del DataFrame 'estudiantes' y aplica el método .describe()
estudiantes["Genero"].describe()
```

```python
# Supongamos que tu DataFrame se llama estudiantes
media_edad = estudiantes["Edad"].mean()

# diferencia respecto a la media
estudiantes["Edad"].map(lambda edad: edad - media_edad)
```
